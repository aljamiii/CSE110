```python
#6 if we separet using comma and space both.
"7, 13, 2, 10, 6, -11, 0"


user=input()
new=""
for i in user:
  if i==",":
    continue
  else:
    new+=i
new=new+" "
#print(new)
newlist=[]
add=""
for j in new:
  if j!=" ":
    add+=j
  elif j==" ":
    newlist.append(int(add))
    add=""
print("My list:",newlist)
max=0
index=0
for k in range(len(newlist)):
  if newlist[k]>max:
    max=newlist[k]
    index=k
print(f"Largest number in the list is {max} which was found at index {index}")
```

```python
#19

list1 = [1,2,2,4,5,5,7,99,200,303,70]
list2 = [1,1,2,3,3,3,4,5,200,500,-5]
newlist=[]
for i in range(len(list1)):
  if newlist==[]:
    newlist.append(list1[i])
  if list1[i] not in newlist:
    newlist.append(list1[i])
new=[]
for j in range(len(list2)):
  if new==[]:
    new.append(list1[j])
  if list2[j] not in new:
    new.append(list2[j])

for k in new:
  if k not in newlist:
    newlist.append(k)
print(newlist)
```

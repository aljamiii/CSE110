```python
#18


sv=int(input())
ev=int(input())
fd=int(input())
sd=int(input())
def function(sv,ev,fd,sd):
  sum=0
  for i in range(sv,ev):
    if i%fd==0 and i%sd==0:
      continue
    elif i%fd==0 or i%sd==0:
      sum+=i
  return sum
function(sv,ev,sd,fd)
```

```python
#20


def bonus(name,income,goal,percent):
  bonus=goal*(percent/100 * income)
  if goal>30:
    bonus=bonus+10000
  elif 20<goal<=30:
    bonus=bonus+5000
  return "{} earned a bonus of {} Taka for {} Goals.".format(name,bonus,goal)
bonus('Luis', 80000, 25, 10)
```

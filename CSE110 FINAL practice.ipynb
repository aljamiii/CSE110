```python
def show_palindrom(num):
  sum=""
  for i in range(1,num):
    sum+=str(i)+" "
  for j in range(num,0,-1):
    sum+=str(j)+" "
  return sum
num=int(input())
show_palindrom(num)

def show_palindromic_triangle(num):
  out=""
  for i in range(1,num+1):
    for spc in range((num-i)*2):
      out+=" "
    out+=show_palindrom(i)
    print(out)
    out=""
show_palindromic_triangle(num)
```

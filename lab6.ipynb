```python
#task8 approach 1

adict="{'Jon':100,'Dan':200,'Rob':300}"
adict=adict.strip("{}")
alist=adict.split(",")
fdict={}
for i in alist:
  blist=i.split(":")
  fdict[blist[0]]=int(blist[1])

newdict={}
for key,value in fdict.items():
  ref=key.strip("'")
  newdict[ref]=value 
  
print(newdict)
```

```python
adict=input()
adict=adict.strip("{}")
alist=adict.split(",")
fdict={}
for i in alist:
  blist=i.split(":")
  fdict[blist[0].strip("'")]=int(blist[1])

sum=0
count=0
for value in fdict.values():
  sum+=value
  count+=1
avg=sum//count
print("Average is",avg)
```

```python
#20

given="i love programming. python is love."
special=['@', '$', '&', '#']
newgiven=""
for i in given:
  if i!=".":
    newgiven+=i
print(newgiven)
word=""
adict={}
sum=0
for j in newgiven:
  if j!=" ":
    word+=j
    sum+=ord(j)
  elif j==" ":
    index=sum%len(special)
    adict[special[index]]=[]
    adict[special[index]].append(word)
    word=""
    sum=0
print(adict)
```

```python
#20

given="i love programming. python is love."
special=['@', '$', '&', '#']
newgiven=""
for i in given:
  if i!=".":
    newgiven+=i
print(newgiven)
word=""
adict={}
sum=0
newgiven=newgiven+" "
for j in newgiven:
  if j!=" ":
    word+=j
    sum+=ord(j)
  elif j==" ":
    index=sum%len(special)
    if special[index] not in adict:
      adict[special[index]]=[]
      adict[special[index]].append(word)
    elif special[index] in adict:
      adict[special[index]].append(word)    
    word=""
    sum=0
print(adict)
```

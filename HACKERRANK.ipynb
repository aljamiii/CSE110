#Given the participants' score sheet for your University Sports Day, you are required to find the runner-up score. You are given  scores. Store them in a list and find the score of the runner-up


```python
n = int(input())
high=0

mylist=[]
for i in range(n):
  mylist.append(int(input()))

for j in range(len(mylist)):
  if high==0:
    high=mylist[j]
  elif mylist[j]>high:
    high=mylist[j]
newlist=[]
for k in mylist:
  if k!=high:
    newlist.append(k)

secondhigh=0
for h in range(len(newlist)):
  if secondhigh==0:
    secondhigh=newlist[h]
  elif newlist[h]>secondhigh:
    secondhigh=newlist[h]
print(secondhigh)
  
```

 Write a Python program to count the number of characters (character frequency) in a string. Go to the editor
Sample String : google.com'
Expected Result : {'g': 2, 'o': 3, 'l': 1, 'e': 1, '.': 1, 'c': 1, 'm': 1}

```python
s=input()
length=len(s)
new=""
for i in range(0,length):

  if new=="":
    new+=s[i]
  else:
    for j in s:
      if j not in new:
        new+=j
#print(new)

for h in new:
  count=0
  for l in range(0,length):
    if h==s[l]:
      count+=1
  print(h,count)
  count=0
```

 Write a Python function that takes a list of words and return the longest word and the length of the longest one.
Sample Output:
Longest word: Exercises
Length of the longest word: 9

```python
s=input()
max=""
min=s
demo=""


for i in s:
  if i!=" ":
    demo+=i
  elif i==" ":
    if len(demo)<len(min):
      min=demo
    elif len(demo)>len(max):
      max=demo
    demo=""
print(f"Longest word:{max}\nLength of longest word:{len(max)}")

  
```

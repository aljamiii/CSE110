#task16

```python
num = int(input("Enter the number of inputs: "))
sum = 0
max = None
min = None
count = 0

while count < num:
    input_num = float(input("Enter a number: "))
    sum += input_num
    if max is None or input_num > max:
        max = input_num
    elif min is None or input_num < min:
        min = input_num
    count += 1

avg = sum / num

print("Maximum: ", max)
print("Minimum: ", min)
print("Average: ", avg)

```
